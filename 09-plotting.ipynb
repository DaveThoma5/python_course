## Plotting in Python

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
time = [0, 1, 2, 3]
position = [0, 100, 24, 56]

plt.plot(time, position)
plt.xlabel("Time (minutes)")
plt.ylabel("Position (cm)")
```

```python
oceania_data = pd.read_csv('data/gapminder_gdp_oceania.csv', index_col='country')
```

```python
print(oceania_data.columns)
years = oceania_data.columns.str.strip('gdpPercap_').astype(int)
```

```python
print(years)
```

```python
oceania_data_mod = oceania_data.copy()
oceania_data_mod.columns = years
```
